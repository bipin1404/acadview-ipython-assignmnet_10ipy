# **QUESTION 2**

```
df.dropna(subset=['Survived'])
df.head()
```

## **NEW COLUMN HAS_CABIN**

```
df['Has_cabin']=np.where(df['Cabin'].isnull(),"NO",'YES')
df.head(8)
```

## **QUESTION 5**

```
df['Cabin'].fillna(0,inplace=True)
df['Age'].fillna(0,inplace=True)
df.head(10)
```

## QUESTION 6

```
df["Categorical_age"]=pd.qcut(df["Age"],1)
df["Categorical_age"].head()
```

```
df["Categorical_Fare"]=pd.qcut(df["Age"],1)
df["Categorical_Fare"].head()
```

```
df.head()
```

# ***TRAIN***(QUESTION 9,10,11)

```
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
reg=LogisticRegression()
```

```
labels = df['Survived']
train1=df.drop(columns=['Survived'],axis=1)
train1.head()
```

```
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
model=reg.fit(x_train,y_train)
model
```

```
print("accuracy of model")
print("score=",model.score(x_test,y_test))
```
